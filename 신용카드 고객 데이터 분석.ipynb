### 소득 수준(Income_Category), 신용 한도(Credit_Limit), 카드 사용 비율(Avg_Utilization_Ratio) 정규화

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 클러스터링에 사용할 변수 선택
cluster_data = data[['Credit_Limit', 'Avg_Utilization_Ratio', 'Income_Category']]

# 데이터 정규화
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)

# 정규화된 데이터 확인
print(cluster_data_scaled[:5])
```

```python
# 클러스터 수에 따른 inertia 계산
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(cluster_data_scaled)
    inertia.append(kmeans.inertia_)

# Elbow Method 시각화
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# K-Means 클러스터링 실행 (최적의 클러스터 수는 4)
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(cluster_data_scaled)

# 클러스터링 결과 확인
print(data[['Credit_Limit', 'Avg_Utilization_Ratio', 'Income_Category', 'Cluster']].head())
```

```python
import seaborn as sns

# 클러스터별 평균값 계산
cluster_summary = data.groupby('Cluster')[['Credit_Limit', 'Avg_Utilization_Ratio', 'Income_Category']].mean()
print(cluster_summary)

# 클러스터별 분포 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Credit_Limit', data=data)
plt.title('Credit Limit by Cluster')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Avg_Utilization_Ratio', data=data)
plt.title('Avg Utilization Ratio by Cluster')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Income_Category', data=data)
plt.title('Income_Category by Cluster')
plt.show()
```

전략: 낮은 수수료와 생활비 할인을 제공하는 카드 추천.
추천 카드: 고급 혜택 카드(포인트 적립, 여행 혜택).
전략: 사용 비율이 낮으므로 혜택을 강화하여 카드 사용 빈도를 증가시키는 전략이 필요.



클러스터 1 (중신용, 저소득, 중간 사용 비율)
특성: 중간 신용 한도, 소득 불명확 또는 저소득층.
전략: 간단한 혜택 구조(저축형, 캐시백 카드)를 추천하여 고객 유지.
추천 카드: 저소득층 맞춤형 혜택 카드(생활비 할인, 캐시백).
전략: 소득 대비 적절한 혜택 제공 및 고객 유지를 위한 적당한 보상 정책.



클러스터 2 (저신용, 중소득, 높은 사용 비율)
특성: 낮은 신용 한도, 높은 카드 사용 비율, 중상위 소득층($60K - $80K).
전략: 신용 한도를 높이거나 저금리 카드로 업그레이드 제안.

```python
from sklearn.model_selection import GridSearchCV
from tqdm import tqdm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# tqdm을 사용한 GridSearchCV 진행 상태 표시
class TqdmGridSearchCV(GridSearchCV):
    def fit(self, X, y=None, **fit_params):
        self._total = len(self.cv_results_.get("params", [])) if hasattr(self, "cv_results_") else 0
        with tqdm(total=self._total, desc="GridSearch Progress") as pbar:
            self.pbar = pbar
            super().fit(X, y, **fit_params)
            self.pbar.close()

# 랜덤 포레스트 하이퍼파라미터 튜닝
param_grid = {
    'n_estimators': [100],
    'max_depth': [10, 20],
    'min_samples_split': [2],
    'min_samples_leaf': [1],
    'class_weight': [class_weights]
}

from sklearn.model_selection import RandomizedSearchCV

random_search = RandomizedSearchCV(RandomForestClassifier(random_state=42), param_grid, n_iter=10, cv=3, scoring='f1_macro', random_state=42)
random_search.fit(X_train_smote, y_train_smote)
```

```python
# 최적의 하이퍼파라미터 출력
print("Best parameters:", random_search.best_params_)

# 최적 모델로 테스트 데이터 평가
best_model = random_search.best_estimator_
y_pred_best = best_model.predict(X_test)

# 결과 출력
print("Classification Report (Best Model):")
print(classification_report(y_test, y_pred_best))
```

```python
from imblearn.combine import SMOTEENN
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# SMOTE + ENN 적용
smote_enn = SMOTEENN(random_state=42)
X_smoteenn, y_smoteenn = smote_enn.fit_resample(X, y)

# 데이터 분포 확인
print("Resampled class distribution (SMOTE+ENN):")
print(pd.Series(y_smoteenn).value_counts())

# 모델 학습 및 평가
model_smoteenn = RandomForestClassifier(random_state=42, class_weight='balanced')
model_smoteenn.fit(X_smoteenn, y_smoteenn)
y_pred_smoteenn = model_smoteenn.predict(X_test)

# 결과 출력
print("Classification Report (SMOTE+ENN):")
print(classification_report(y_test, y_pred_smoteenn))
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, f1_score

# 모델 정의
models = {
    "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced'),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'),
    "LightGBM": LGBMClassifier(random_state=42)
}

# 각 모델 학습 및 평가
best_model = None
best_f1_macro = 0

for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_smoteenn, y_smoteenn)
    y_pred = model.predict(X_test)
    
    # 성능 평가
    print(f"Classification Report for {name}:")
    print(classification_report(y_test, y_pred))
    
    # F1-Score 비교
    f1_macro = f1_score(y_test, y_pred, average='macro')
    print(f"F1 Macro for {name}: {f1_macro}")
    
    if f1_macro > best_f1_macro:
        best_f1_macro = f1_macro
        best_model = model

print(f"\nBest Model: {best_model} with F1 Macro: {best_f1_macro}")
```
